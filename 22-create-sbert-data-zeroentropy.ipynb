```python
from sentence_transformers import CrossEncoder
model = CrossEncoder("zeroentropy/zerank-1-small", trust_remote_code=True)
```

```python
from datasets import load_dataset
train_dataset = load_dataset("sentence-transformers/all-nli", "pair-score", split="train")
```

```python
df = train_dataset.to_pandas()
```

```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
df.head(20).style.background_gradient(cmap='coolwarm')
```

```python
df["sentence1"].nunique()
```

```python
df["sentence2"].nunique()
```

```python
dft = df.head(20).copy()
```

```python
pairs = [(s1, s2) 
           for s1, s2 in zip(dft["sentence1"], dft["sentence2"])]
```

```python
%%time
scores = model.predict(pairs)
```

```python
dft["scores"] = scores
```
